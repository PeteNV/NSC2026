```python
from ultralytics import YOLO

model = YOLO('v1/v1.pt')

class_names = {
    0: 'air_conditioner',
    1: 'air_fryer',
    2: 'blender',
    3: 'computer',
    4: 'electric_kettle',
    5: 'fan',
    6: 'induction_cooker',
    7: 'light',
    8: 'microwave',
    9: 'oven',
    10: 'refrigerator',
    11: 'rice_cooker',
    12: 'screen',
    13: 'toaster',
    14: 'washing_machine'
}

# Manually override the model's internal names dictionary.
model.model.names = class_names

# Export to CoreML
# imgsz=640: Standard for YOLOv8 mobile deployment.
# nms=True: Box-filtering logic into the model
model.export(
    format='coreml', 
    imgsz=640, 
    nms=True
)

print("Export complete")
```

```python
# Testing via CoreML model and webcam
from ultralytics import YOLO
import cv2

class_names = {
    0: 'air_conditioner',
    1: 'air_fryer',
    2: 'blender',
    3: 'computer',
    4: 'electric_kettle',
    5: 'fan',
    6: 'induction_cooker',
    7: 'light',
    8: 'microwave',
    9: 'oven',
    10: 'refrigerator',
    11: 'rice_cooker',
    12: 'screen',
    13: 'toaster',
    14: 'washing_machine'
}

# Set NV's mac webcam
cap = cv2.VideoCapture(1)

model = YOLO("v1/v1_patched.mlpackage")

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Run inference on the frame
    results = model(frame)
    for result in results:
        result.names = class_names

    # Visualize the results on the frame
    annotated_frame = results[0].plot()

    # Display the frame
    cv2.imshow("CoreML YOLO Detection", annotated_frame)

    # Break loop with 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import torch
from ultralytics import YOLO

original_weights = 'v1/v1.pt'
patched_weights = 'v1/v1_patched.pt'

class_names = {
    0: 'air_conditioner',
    1: 'air_fryer',
    2: 'blender',
    3: 'computer',
    4: 'electric_kettle',
    5: 'fan',
    6: 'induction_cooker',
    7: 'light',
    8: 'microwave',
    9: 'oven',
    10: 'refrigerator',
    11: 'rice_cooker',
    12: 'screen',
    13: 'toaster',
    14: 'washing_machine'
}

ckpt = torch.load(original_weights, weights_only=False)

ckpt['model'].names = class_names

torch.save(ckpt, patched_weights)
print(f"Saved patched weights to {patched_weights}")

model = YOLO(patched_weights)

# Export to CoreML
model.export(
    format='coreml', 
    imgsz=640, 
    nms=True
)

print("Export complete")
```
